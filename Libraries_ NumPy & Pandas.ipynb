Question:1 Load the given retail sales dataset into a Pandas DataFrame and display the first 5 rows along with the
dataset information to understand its structure. **bold text**

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('/retail_dataset  (1).csv')

# Display afirst 5 rows
print("First 5 Rows of Dataset:")
display(df.head())

# Display dataset information
print("\nDataset Information:")
df.info()
```

Question:2 Inspect the dataset and identify missing values in each column. Display the total number of missing
values present in the dataset.

```python
# Check missing values in each column
missing_values = df.isnull().sum()

print("Missing Values in Each Column:")
print(missing_values)

# Total missing values
total_missing = missing_values.sum()

print("\nTotal Missing Values in Dataset:", total_missing)
```

Question:3. Replace missing values in the Quantity and Price columns with their respective mean values using
Pandas.

```python
# Replace missing values with mean
df['Quantity'].fillna(df['Quantity'].mean(), inplace=True)
df['Price'].fillna(df['Price'].mean(), inplace=True)

print("Missing values replaced successfully.")

# Verify
print(df[['Quantity', 'Price']].isnull().sum())
```

Question:4. Remove all rows where the Product Category or Region column contains missing values.


```python
# Remove rows with missing Product Category or Region
df = df.dropna(subset=['Product Category', 'Region'])

print("Rows with missing Product Category or Region removed.")
print("\nRemaining Missing Values:")
print(df.isnull().sum())
```

Question:5. Create a new column called Revenue using NumPy, where:

Revenue = Quantity × Price

Display the updated dataset with the new column.

```python
import numpy as np

# Create Revenue column
df['Revenue'] = np.multiply(df['Quantity'], df['Price'])

# Display updated dataset
display(df.head())
```

Question:6. Using NumPy, calculate the total revenue generated from all transactions in the dataset.

```python
# Total revenue
total_revenue = np.sum(df['Revenue'])

print("Total Revenue Generated:", total_revenue)
```

Question:7. Group the dataset by Product Category and calculate the total revenue generated by each category.

```python
# Revenue by Product Category
category_revenue = df.groupby('Product Category')['Revenue'].sum()

print("Total Revenue by Product Category:")
print(category_revenue)
```

Question:8. Based on the revenue calculated for each product category, identify the top 3 categories with the
highest revenue and the bottom 3 categories with the lowest revenue.

```python
# Top 3 categories
top_3 = category_revenue.sort_values(ascending=False).head(3)

# Bottom 3 categories
bottom_3 = category_revenue.sort_values().head(3)

print("Top 3 Categories by Revenue:")
print(top_3)

print("\nBottom 3 Categories by Revenue:")
print(bottom_3)
```

Question:9. Group the dataset by Region and calculate the total revenue generated in each region. Identify the
region with the highest revenue and the region with the lowest revenue.

```python
# Revenue by Region
region_revenue = df.groupby('Region')['Revenue'].sum()

print("Revenue by Region:")
print(region_revenue)

# Highest revenue region
highest_region = region_revenue.idxmax()
highest_value = region_revenue.max()

# Lowest revenue region
lowest_region = region_revenue.idxmin()
lowest_value = region_revenue.min()

print("\nHighest Revenue Region:", highest_region, "-", highest_value)
print("Lowest Revenue Region:", lowest_region, "-", lowest_value)
```

Question:10. Extract the month from the Date column using Pandas and group the dataset by month to calculate the
total monthly revenue. Also, calculate the mean, median, and standard deviation of the Revenue column
using NumPy.

```python
# Convert Date column to datetime
df['Date'] = pd.to_datetime(df['Date'], format='%d-%m-%Y')

# Extract month
df['Month'] = df['Date'].dt.month

# Monthly revenue
monthly_revenue = df.groupby('Month')['Revenue'].sum()

print("Monthly Revenue:")
print(monthly_revenue)

# Revenue statistics using NumPy
mean_revenue = np.mean(df['Revenue'])
median_revenue = np.median(df['Revenue'])
std_revenue = np.std(df['Revenue'])

print("\nRevenue Statistics:")
print("Mean Revenue:", mean_revenue)
print("Median Revenue:", median_revenue)
print("Standard Deviation of Revenue:", std_revenue)
```
